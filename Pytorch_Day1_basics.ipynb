# Tensor Range

```python
zero_to_one = torch.range(start=0,end=10,step=1).type(torch.int32)
zero_to_one, zero_to_one.dtype
```

### Another tensor with same shape with one tensor

```python
ten_zeros = torch.zeros_like(input=zero_to_one)
ten_zeros
```

```python
ten_one = torch.ones_like(input=zero_to_one)
ten_one
```
